```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./graph_scores.csv')
df.index = range(10,101,10)
df
```

```python
line_width = 4
marker_size = 10
```

```python
plt.figure(figsize=(30,10))
plt.subplots_adjust(wspace=0.25)

plt.subplot(1, 2, 1)
plt.title('$NDCG_K$', fontsize=30, pad=20)
plt.plot(df['normal_ndcg'], label='KGAT', linewidth=line_width, marker='o', markersize=marker_size, c='cornflowerblue')
plt.plot(df['normal_embedding_ndcg'], label='KGAT with Text Embedding', linewidth=line_width, marker='o', markersize=marker_size, c='royalblue')
plt.xticks(range(10,101,10), fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('$K$', fontsize=25)
plt.ylabel('$NDCG_K$', fontsize=25)
plt.legend(loc='lower right', fontsize=20)
plt.grid(linestyle='--', alpha=0.5)

plt.subplot(1, 2, 2)
plt.title('$Recall_K$', fontsize=30, pad=20)
plt.plot(df['normal_recall'], label='KGAT', linewidth=line_width, marker='o', markersize=marker_size, c='cornflowerblue')
plt.plot(df['normal_embedding_recall'], label='KGAT with Text Embedding', linewidth=line_width, marker='o', markersize=marker_size, c='royalblue')
plt.xticks(range(10,101,10), fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('$K$', fontsize=25)
plt.ylabel('$Recall_K$', fontsize=25)
plt.legend(loc='lower right', fontsize=20)
plt.grid(linestyle='--', alpha=0.5)

# plt.show()
plt.savefig('figure_1.pdf', dpi=1200)
```
